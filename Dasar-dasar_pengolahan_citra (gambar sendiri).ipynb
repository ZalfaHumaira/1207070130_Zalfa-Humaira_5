```python
import matplotlib.pyplot as plt

from skimage import data
from skimage.io import imread
from skimage.color import rgb2gray
from skimage.util import invert
from PIL import ImageEnhance
import numpy as np
import cv2
```

```python
#percobaan 1 : crop image

#load image dari file 
citraImage= cv2.imread('citra.jpg')
#mengubah gambar menjadi RGB
citraImage = cv2.cvtColor(citraImage, cv2.COLOR_BGR2RGB)

tootImage = cv2.imread('toot.jpg')
tootImage = cv2.cvtColor(tootImage, cv2.COLOR_BGR2RGB)

#menyimpan citra yang sudah di potong
citraCropped = citraImage.copy()
citraCropped = citraImage[0:256,64:320]

tootCropped = tootImage.copy()
tootCropped = tootCropped[64:256,128:320]

#menampilkan citra asli dan hasil crop dalam 1 figure
fig, axes = plt.subplots(2, 2, figsize=(6, 6))
ax = axes.ravel()
ax[0].imshow(citraImage)
ax[0].set_title("Citra Input 1")
ax[1].imshow(tootImage, cmap='gray')
ax[1].set_title("Citra Input 2")
ax[2].imshow(citraCropped)
ax[2].set_title("Citra Output 1")
ax[3].imshow(tootCropped, cmap='gray')
ax[3].set_title("Citra Output 2")
```

```python
#membalikan citra yang tersimpan
inv  = invert( citraCropped ) 
print ( 'Shape Input : ' ,  citraCropped . shape ) 
print ( ' Shape Output : ' , inv . shape )
#menampilkan gambar asli dan hasil invert beserta histogramnya dalam 1 figure
fig ,axes  =  plt . subplots ( 2 ,  2 ,  figsize = ( 6 ,  6 )) 
ax  =  axes .ravel ()
ax[0].imshow ( citraCropped )
ax[0].set_title ( "Input Citra" )
#menampilkan histogram gambar asli
ax[1].hist ( citraCropped .ravel ( ), bins = 256 ) 
ax[1].set_title ( 'Masukan histogram' ) 
ax[2].imshow ( inv ) 
ax[2].set_title ( 'Citra Output (Gambar Terbalik)' )
#menampilkan histogram hasil invert
ax[3].hist ( inv . ravel (),  bins = 256 ) 
ax[3].set_title ( 'Keluaran histogram' )
```

```python
#membuat brightness pada output Toot
#mengubah tipe data menjadi float
copytoot = tootCropped.copy().astype(float)

#menambahkan parameter 100 pada setiap elemen aray copyCamera untuk brightness
for i in range (copytoot.shape[0]):
    for j in range (copytoot.shape[1]):
       copytoot[i,j] += 100
copytoot[copytoot > 255] = 255
copytoot = copytoot.astype(np.uint8)

#menampilkan gambar asli dan hasil brightness berikut histogramnya dalam 1 figure
fig, axes = plt.subplots(2,2, figsize=(6,6))
ax = axes.ravel()
ax[0].imshow(tootCropped, cmap ='gray')
ax[0].set_title("Citra Input")
ax[1].hist(tootCropped.ravel(), bins=256)
ax[1].set_title("Histogram Input")
ax[2].imshow(copytoot, cmap ='gray')
ax[2].set_title("Citra Output (Brightness)")
ax[3].hist(copytoot.ravel(), bins=256)
ax[3].set_title("Histogram Input")
```
