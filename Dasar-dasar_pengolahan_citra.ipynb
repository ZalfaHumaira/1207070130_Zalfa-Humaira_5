```python
import matplotlib.pyplot as plt

from skimage import data
from skimage.io import imread
from skimage.color import rgb2gray
from skimage.util import invert
import numpy as np
```

```python
#percobaan 1 : crop image

#load gambar dari skimage
astronautImage = data.astronaut()
cameraImage = data.camera()

#membuat fungsi crop gambar astronaut
astroCropped = astronautImage.copy()
astroCropped = astroCropped[0:256,64:320]

#membuat fungsi crop gambar camera
cameraCropped = cameraImage.copy()
cameraCropped = cameraCropped[64:256,128:320]

#menampilkan gambar asli dan hasil crop dalam 1 figure menggunakan subplot
fig, axes = plt.subplots(2, 2, figsize=(6, 6))
ax = axes.ravel()
ax[0].imshow(astronautImage)
ax[0].set_title("Citra Input 1")
ax[1].imshow(cameraImage, cmap='gray')
ax[1].set_title("Citra Input 2")
ax[2].imshow(astroCropped)
ax[2].set_title("Citra Output 1")
ax[3].imshow(cameraCropped, cmap='gray')
ax[3].set_title("Citra Output 2")
```

```python
#membalikan citra yang tersimpan pada astroCropped
inv  = invert( astroCropped ) 
print ( 'Shape Input : ' ,  astroCropped . shape ) 
print ( ' Shape Output : ' , inv . shape )

#menampilkan gambar asli dan hasil invert beserta histogramnya dalam 1 figure
fig ,axes  =  plt . subplots ( 2 ,  2 ,  figsize = ( 6 ,  6 )) 
ax  =  axes .ravel ()
ax[0].imshow ( astroCropped ) 
ax[0].set_title ( "Input Citra" )
#menampilkan histogram gambar asli
ax[1].hist ( astroCropped .ravel ( ), bins = 256 ) 
ax[1].set_title ( 'Masukan histogram' ) 
ax[2].imshow ( inv ) 
ax[2].set_title ( 'Citra Output (Gambar Terbalik)' )
#menampilkan histogram hasil invert
ax[3].hist ( inv . ravel (),  bins = 256 )
ax[3].set_title ( 'Keluaran histogram' )
```

```python
#membuat brightness pada output camera
#mengubah tipe data menjadi float
copyCamera = cameraCropped.copy().astype(float)

#mengambil nilai m1 dan n1 dari copyCamera shape dan dibuat aray dg ukuran yang sama
m1,n1 = copyCamera.shape
output1 = np.empty([m1,n1])

#menambahkan parameter 100 pada setiap elemen aray copyCamera untuk brightness
for baris in range(0,m1-1):
    for kolom in range(0,n1-1):
        a1 = baris
        b1 = kolom
        output1[a1,b1] = copyCamera[baris, kolom] + 100

#menampilkan gambar asli dan hasil brightness berikut histogramnya dalam 1 figure
fig, axes = plt.subplots(2,2, figsize=(6,6))
ax = axes.ravel()
ax[0].imshow(cameraCropped, cmap ='gray')
ax[0].set_title("Citra Input")
ax[1].hist(astroCropped.ravel(), bins=256)
ax[1].set_title("Histogram Input")
ax[2].imshow(output1, cmap ='gray')
ax[2].set_title("Citra Output (Brightness)")
ax[3].hist(output1.ravel(), bins=256)
ax[3].set_title("Histogram Input")
```
